### 将每一列中，缺失值较多的列进行删除

```python
def filter_col_by_nan(df,ratio=0.05):
    cols=[]
    for col in df.columns:
        if df[col].isna().mean()>=(1-ratio):
            cols.append(col)
        return cols
```
